### Extracting the records of the movie which have the IMDB ratings greater than 7 and Meta Score 50.

```python
meta = df.loc[(df['IMDB_Rating'] >=7 ) & (df['Meta_score'] >= 50)]
metascore = meta.loc[: , ['Title' ,'Duration' ,'IMDB_Rating'  ,'Meta_score', 'Genre']]
metascore.head(5)
```

### Visualizing the plot for Movie of different duration which got different certificates

```python
plt.figure(figsize=(10,6))
genre = df.loc[: , ['Certificate' ,'Title' , 'Duration']].sample(15)
sns.set_theme(style="whitegrid")
sns.barplot(x="Duration", y="Title", data=genre , hue ='Certificate' )
```

### Extracting the records for the Movie Title along with its genre.

```python
tit = df.loc[: , ['Title' , 'Genre']].head(20)
tit
```

```python
fig = plt.figure(figsize=(15,8))
x=tit["Title"].to_numpy()
y=tit["Genre"].to_numpy()
plt.bar(x,y,color="y")
plt.xticks(x,rotation=45)
plt.title("Movie and the Genre")
plt.ylabel("Genre")
plt.xlabel("Title")
plt.show()

```

### Barplot to see movie and the gross income

```python
fig = plt.figure(figsize=(10,8))
Title_gross = df.loc[: ,['Gross' , 'Title' , 'Released_Year' ]].sample(10)
sns.barplot(x = "Gross", y= 'Title' ,data=Title_gross , palette = 'winter')
```

### Replot to show the which movie has grossed the highest and with which certification.

```python
fig = plt.figure(figsize=(10,4))
Gross_title = df.loc[: ,['Gross' , 'Title' ,'Certificate' ]].sample(20)
sns.scatterplot(x="Gross", y="Title", hue = "Certificate", data = Gross_title)
```

### Plotting Voilin plot for Meta score and IMDB Ratings.

```python
Metaimdb = df.loc[:,['Meta_score' , 'IMDB_Rating']]
Metaimdb.sample(20)
```

```python
plt.figure(figsize= (10, 5))
sns.violinplot(x = "IMDB_Rating", y = "Meta_score" ,data=Metaimdb)
```

```python
fig = plt.figure(figsize=(10,6))
sns.distplot(df[['IMDB_Rating', 'Meta_score']])
fig.legend(labels=['IMDB Ratings','Meta_score'])
plt.show()
```
